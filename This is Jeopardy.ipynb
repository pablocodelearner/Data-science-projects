2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
def questions_containing_words(list_of_words):
    filtered_df = df[df.question.apply(lambda x: word in x for word in list_of_words).all(axis=1)]
    return filtered_df

df_filtered_1 = questions_containing_words(['King', 'England'])
display(df_filtered_1.head(10))
print(df_filtered_1.info())
````

3. Test your original function with a few different sets of words to try to find some ways your function breaks. Edit your function so it is more robust.

   For example, think about capitalization. We probably want to find questions that contain the word `"King"` or `"king"`.
   
   You may also want to check to make sure you don't find rows that contain substrings of your given words. For example, our function found a question that didn't contain the word `"king"`, however it did contain the word `"viking"` &mdash; it found the `"king"` inside `"viking"`. Note that this also comes with some drawbacks &mdash; you would no longer find questions that contained words like `"England's"`.

````python
def questions_containing_words_2(list_of_words):
    filtered_df = df[df.question.apply(lambda x: word in x or word.lower() in x for word in \
    list_of_words).all(axis=1)]
    return filtered_df

df_filtered_2 = questions_containing_words_2(['King', 'England'])

display(df_filtered_2.head(10))
print(df_filtered_2.info())
````

4. We may want to eventually compute aggregate statistics, like `.mean()` on the `" Value"` column. But right now, the values in that column are strings. Convert the`" Value"` column to floats. If you'd like to, you can create a new column with float values.

   Now that you can filter the dataset of question, use your new column that contains the float values of each question to find the "difficulty" of certain topics. For example, what is the average value of questions that contain the word `"King"`?
   
   Make sure to use the dataset that contains the float values as the dataset you use in your filtering function.

````python
# Adding a new column called 'float_value' for performing aggregate functions:

str_to_float = lambda x: float(("".join(x.split(',')).strip('$'))) if type(x) == str else x

df['value'] = df.value.apply(lambda x: 0.0 if x == 'None' else x)

df['float_value'] = df.value.apply(str_to_float)

# Looking for mean values of questions containing different words from a list:

def mean_value_for_questions_containing(list_of_words):
    means_dict = dict()
    for i in list_of_words:
        means_dict['Mean value for questions containing the word {}'.format(i)] = \
        round(questions_containing_words([i]).float_value.mean(), 2)
    return means_dict
        
some_words_list = ['king', 'basketball', 'movie', 'president', 'French'] 
some_mean_values = mean_value_for_questions_containing(some_words_list)

for i in some_mean_values:
    print(i + ": " + str(some_mean_values[i]))
````

5. Write a function that returns the count of unique answers to all of the questions in a dataset. For example, after filtering the entire dataset to only questions containing the word `"King"`, we could then find all of the unique answers to those questions. The answer "Henry VIII" appeared 55 times and was the most common answer.

````python
def unique_answers_count(df):
    return (df.groupby('answer').answer.count().to_frame()).rename(columns={'answer': 'answer',\
                                                                         'answer': 'count'})

king_answers = unique_answers_count(questions_containing_words_2(['King']))

king_answers.reset_index(inplace=True)

henry_viii_count = questions_containing_words_2(['King']).answer.apply(lambda x: x if x == 'Henry VIII'\
                                                        else None).count()

print(king_answers.head(15))
print("\n" + "'Henry VIII' answers in questions containing the word 'King' (or 'king'): " + \
      str(henry_viii_count))
````
